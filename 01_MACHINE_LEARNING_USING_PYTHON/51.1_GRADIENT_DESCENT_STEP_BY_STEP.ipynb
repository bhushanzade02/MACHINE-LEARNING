```python
#gradient descent for m and b
b = -100
m = 60
lr = 0.01
iterations = 100


for i in range(iterations):
    loss_function = y - (m * X[:, 0] + b)
    loss_slope_b = -2 * np.sum( y - (m * X[:, 0] + b))  
    loss_slope_m = -2 * np.sum(X[:, 0] * ( y - (m * X[:, 0] + b)))  

  
    b = b - lr * loss_slope_b
    m = m - lr * loss_slope_m


print(m, b)
```
