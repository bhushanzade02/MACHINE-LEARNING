```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import SelectKBest, chi2

# Define preprocessing steps
categorical_features = ['Sex', 'Embarked']
numerical_features = ['Age', 'Fare']

# Define transformers
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
numerical_transformer = SimpleImputer(strategy='mean')

# ColumnTransformer to handle both categorical and numerical data
preprocessor = ColumnTransformer([
    ('num', numerical_transformer, numerical_features),
    ('cat', categorical_transformer, categorical_features)
])

# Feature selection
feature_selector = SelectKBest(score_func=chi2, k=8)

# Decision Tree Classifier
classifier = DecisionTreeClassifier()

# Pipeline
pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('feature_selection', feature_selector),
    ('classifier', classifier)  # Ensure this is an instance, not a class
])

# Fit the pipeline
pipe.fit(X_train, y_train)  # Make sure X_train and y_train are correctly preprocessed
```
