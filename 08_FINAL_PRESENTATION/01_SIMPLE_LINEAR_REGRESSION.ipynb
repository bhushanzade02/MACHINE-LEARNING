```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(R"placement_cgpa_pcakge.csv")
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'],label="CGPA VS PACKAGE")
plt.xlabel("cgpa")
plt.ylabel("package")
plt.legend()
plt.show()
```

```python
X=df.iloc[:,0:1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

## USING SCIKIT LEARN

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
```

```python
plt.scatter(df['cgpa'], df['package'], color='blue', label='Actual Data')
plt.plot(X_train, lr.predict(X_train), color='red', label='Regression Line')
plt.xlabel("CGPA")  
plt.ylabel("Package (LPA)")  
plt.title("CGPA vs. Package - Linear Regression Using Sklearn")  
plt.legend()  
plt.show()
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,root_mean_squared_error
print("R2 SCORE :",r2_score(y_test,y_pred))
print("MEAN ABSOLUTE ERROR :",mean_absolute_error(y_test,y_pred))
print("MEAN SQUARED ERROR  :",mean_squared_error(y_test,y_pred))
print("ROOT MEAN SQUARED ERROR  :",root_mean_squared_error(y_test,y_pred))
```

## ALGONIX IMPLEMENTATION 

```python
df=pd.read_csv(R"placement_cgpa_pcakge.csv")
X=df.iloc[:,0].values
y=df.iloc[:,1].values
from sklearn.model_selection import train_test_split
X_train ,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from algonix.models import SimpleLinearRegression
```

```python
lr=SimpleLinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
```

```python
plt.scatter(df['cgpa'], df['package'], color='blue', label='Actual Data')
plt.plot(X_train, lr.predict(X_train), color='red', label='Regression Line')
plt.xlabel("CGPA")  
plt.ylabel("Package (LPA)")  
plt.title("CGPA vs. Package - Linear Regression Custom Implementation")  
plt.legend()  
plt.show()
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,root_mean_squared_error
print("R2 SCORE :",r2_score(y_test,y_pred))
print("MEAN ABSOLUTE ERROR :",mean_absolute_error(y_test,y_pred))
print("MEAN SQUARED ERROR  :",mean_squared_error(y_test,y_pred))
print("ROOT MEAN SQUARED ERROR  :",root_mean_squared_error(y_test,y_pred))
```
