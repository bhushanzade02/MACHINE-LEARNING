# ex tp

```python
import matplotlib.pyplot as plt

sns.distplot(df[df['balcony']==1]['price'],hist=False,)
sns.distplot(df[df['balcony']==2]['price'],hist=False,)
sns.distplot(df[df['balcony']==3]['price'],hist=False,)

plt.legend()
plt.title('price distribution by no of balconies')
plt.xlabel('price')
plt.ylabel('density')
plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create KDE plots with labels
sns.kdeplot(df[df['balcony']==1]['price'], label='1 Balcony')
sns.kdeplot(df[df['balcony']==2]['price'], label='2 Balconies')
sns.kdeplot(df[df['balcony']==3]['price'], label='3 Balconies')

# Add legend manually
plt.legend(title="Number of Balconies")

# Labels and title
plt.title("Price Distribution by Number of Balconies")
plt.xlabel("Price")
plt.ylabel("Density")

# Show plot
plt.show()

```
