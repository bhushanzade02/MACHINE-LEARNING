```python
from sklearn.datasets import make_regression
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
X,y=make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20,random_state=13)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(X,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
from sklearn.linear_model import Ridge
```

```python
Rr=Ridge(alpha=10)
Rr.fit(X,y)
print(Rr.coef_)
print(Rr.intercept_)
```

```python
Rr1=Ridge(alpha=100)
Rr1.fit(X,y)
print(Rr.coef_)
print(Rr.intercept_)
```

```python
plt.plot(X,y,'b.')
plt.plot(X,lr.predict(X),'r')
plt.plot(X,Rr.predict(X),'g')
plt.plot(X,Rr1.predict(X),'b')
```
