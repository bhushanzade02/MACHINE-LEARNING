# using heart diseases dataset 

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"D:\fake\heart.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
X=df.iloc[:,0:-1]
```

```python
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf1=LogisticRegression()
clf2=DecisionTreeClassifier()
```

```python
clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
```

```python
y_pred1=clf1.predict(X_test)
y_pred2=clf2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("ACcuracy score of Logistc regression",accuracy_score(y_test,y_pred1))
print("Accuracy score of Decision tree",accuracy_score(y_test,y_pred2))
```

```python
print("Logistic Regression Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,2)))
```

```python

print("Decision Tree Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred2),columns=list(range(0,2)))
```

```python
result=pd.DataFrame()
result["ACtual"]=y_test
result["predisctiokn by Logistic "]=y_pred1
result["Prediction by Decsiosn Tree"]=y_pred2
result.sample(10)
```

```python
from sklearn.metrics import recall_score,f1_score,precision_score,confusion_matrix
```

```python
print("Confusion Matrix for Logistic Regression")
print("-"*50)
cdf=pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,2)))
print(cdf)
print("-"*50)
print("Recall score:",recall_score(y_test,y_pred1))
print("F1 SCore     ",f1_score(y_test,y_pred1))
print("Precisiom    ",precision_score(y_test,y_pred1))
```
